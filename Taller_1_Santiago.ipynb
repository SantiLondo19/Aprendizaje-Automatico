```python
# Modelo con las variables más correlacionadas
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# Convertir atributos categóricos a columnas one-hot
df_encoded = pd.get_dummies(df, columns=['protocol_type', 'service', 'flag', 'class'])

# Seleccionar las variables más correlacionadas (positiva y negativamente) con 'class_numerico'
selected_features = [
    'same_srv_rate', 'dst_host_srv_count', 'dst_host_same_srv_rate',
    'logged_in', 'srv_diff_host_rate', 'count', 'srv_serror_rate', 'serror_rate',
    'dst_host_serror_rate', 'dst_host_srv_serror_rate'
]

# Dividir los datos en atributos (X) y la columna objetivo (y)
X = df_encoded[selected_features]
y = df_encoded['class_numerico']

# Dividir los datos en conjuntos de entrenamiento y pruebas
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Estandarizar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear una instancia de LogisticRegression y entrenar el modelo
logreg_model = LogisticRegression()
logreg_model.fit(X_train_scaled, y_train)

# Realizar la predicción en el conjunto de pruebas
y_test_pred = logreg_model.predict(X_test_scaled)

# Calcular la precisión del modelo en el conjunto de pruebas
accuracy = accuracy_score(y_test, y_test_pred)
print("Precisión del modelo en el conjunto de pruebas:", accuracy)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Reemplazar los signos de interrogación (?) por valores nulos (NaN)
df.replace('?', pd.NA, inplace=True)

# Convertir atributos categóricos a columnas one-hot
df_encoded = pd.get_dummies(df, columns=['protocol_type', 'service', 'flag', 'class'])


# Dividir los datos en conjuntos de entrenamiento y pruebas
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Estandarizar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear una instancia de LogisticRegression y entrenar el modelo
logreg_model = LogisticRegression()
logreg_model.fit(X_train_scaled, y_train)

# Realizar la predicción en el conjunto de pruebas
y_test_pred = logreg_model.predict(X_test_scaled)

# Calcular la precisión del modelo en el conjunto de pruebas
accuracy_test = accuracy_score(y_test, y_test_pred)
print("Precisión del modelo en el conjunto de pruebas:", accuracy_test)

# Calcular la matriz de confusión
confusion = confusion_matrix(y_test, y_test_pred)

# Graficar la matriz de confusión
plt.figure(figsize=(8, 6))
plt.imshow(confusion, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Matriz de Confusión")
plt.colorbar()
tick_marks = np.arange(len(['Normal', 'Anomalía']))
plt.xticks(tick_marks, ['Normal', 'Anomalía'], rotation=45)
plt.yticks(tick_marks, ['Normal', 'Anomalía'])
plt.ylabel('Etiqueta real')
plt.xlabel('Etiqueta predicha')
plt.show()
print("Matriz de Confusión - Datos Validación:")
print(confusion)
```

```python
class_report = classification_report(y_test, y_test_pred, target_names=['Normal', 'Anomalía'])
print("Classification Report:\n", class_report)
```

Precisión (Precision): Para ambas clases, "Normal" y "Anomalía", la precisión es 1.00 (100%). Esto significa que todas las predicciones positivas que el modelo hace son realmente verdaderas positivas. En otras palabras, el modelo no comete errores en términos de predicciones positivas.

Recall (Sensibilidad o Tasa de Verdaderos Positivos): Para ambas clases, el recall también es 1.00 (100%). Esto indica que el modelo es capaz de identificar y capturar todos los casos reales de ambas clases en el conjunto de pruebas. No se pierden ni casos de "Normal" ni casos de "Anomalía".

F1-score: Dado que tanto la precisión como el recall son 1.00 para ambas clases, el F1-score también es 1.00 (100%) para ambas clases. Esto refuerza la idea de que el modelo está realizando perfectamente en la clasificación, ya que el F1-score es una medida de equilibrio entre precisión y recall.

Support: El "support" indica el número real de ejemplos de cada clase en el conjunto de pruebas. Hay 17,709 ejemplos de la clase "Normal" y 20,083 ejemplos de la clase "Anomalía".

```python
# Punto 20
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, auc
# Calcular la curva ROC
y_test_prob = logreg_model.predict_proba(X_test_scaled)[:, 1]

fpr, tpr, thresholds_roc = roc_curve(y_test, y_test_prob)
roc_auc = roc_auc_score(y_test, y_test_prob)

# Calcular la curva PR
precision, recall, thresholds_pr = precision_recall_curve(y_test, y_test_prob)
pr_auc = auc(recall, precision)

# Graficar las curvas ROC y PR
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, color='blue', label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(recall, precision, color='blue', label='PR curve (area = {:.2f})'.format(pr_auc))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Curva PR')
plt.legend(loc='lower left')

plt.tight_layout()
plt.show()
```
