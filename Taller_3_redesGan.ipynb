```python
# Importar bibliotecas necesarias
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Reshape, Conv2D, LeakyReLU
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

```

```python
# Cargar y preprocesar el conjunto de datos MNIST
(x_train, _), (_, _) = mnist.load_data()
x_train = x_train.astype(np.float32) / 255.0
x_train = (x_train - 0.5) * 2.0  # Escalar a valores entre -1 y 1
x_train = np.expand_dims(x_train, axis=-1)  # Agregar dimensiÃ³n de canales

# TamaÃ±o del espacio latente
latent_dim = 100

# Construir el generador
generator = Sequential([
    Dense(128, activation='relu', input_dim=latent_dim),
    Dense(784, activation='tanh'),  # 28x28 para MNIST
    Reshape((28, 28, 1))
])

# Construir el discriminador
discriminator = Sequential([
    Conv2D(64, (3, 3), strides=(2, 2), padding='same', input_shape=(28, 28, 1)),
    LeakyReLU(alpha=0.2),
    Conv2D(128, (3, 3), strides=(2, 2), padding='same'),
    LeakyReLU(alpha=0.2),
    Flatten(),
    Dense(1, activation='sigmoid')
])

# Configurar el discriminador para que no sea entrenable durante el entrenamiento de la GAN
discriminator.compile(loss='binary_crossentropy', optimizer='adam')
discriminator.trainable = False
visualize_every = 136
# Construir la GAN combinando el generador y el discriminador
gan_input = tf.keras.layers.Input(shape=(latent_dim,))
x = generator(gan_input)
gan_output = discriminator(x)
gan = tf.keras.models.Model(gan_input, gan_output)
gan.compile(loss='binary_crossentropy', optimizer='adam')

# FunciÃ³n para entrenar la GAN
def train_gan(gan, generator, discriminator, x_train, latent_dim, epochs=10000, batch_size=64):
    d_losses = []  # Lista para almacenar pÃ©rdidas del discriminador
    g_losses = []  # Lista para almacenar pÃ©rdidas del generador
    for epoch in range(epochs):
        # Entrenar el discriminador
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        fake_images = generator.predict(noise)
        real_images = x_train[np.random.randint(0, x_train.shape[0], batch_size)]
        labels_real = np.ones((batch_size, 1))
        labels_fake = np.zeros((batch_size, 1))
        d_loss_real = discriminator.train_on_batch(real_images, labels_real)
        d_loss_fake = discriminator.train_on_batch(fake_images, labels_fake)
        d_loss = 0.5 * (d_loss_real + d_loss_fake)

        # Entrenar el generador
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        labels_gan = np.ones((batch_size, 1))
        g_loss = gan.train_on_batch(noise, labels_gan)

        d_losses.append(d_loss)
        g_losses.append(g_loss)

        # Imprimir el progreso
        if epoch % 100 == 0:
            print(f"Ã‰poca: {epoch}, D Loss: {d_loss}, G Loss: {g_loss}")

        # Guardar imÃ¡genes generadas en algunas Ã©pocas
        if epoch % visualize_every == 0:
            save_generated_images(generator, epoch)
            print(f"Ã‰poca: {epoch},")
    return d_losses, g_losses

# FunciÃ³n para guardar imÃ¡genes generadas
def save_generated_images(generator, epoch, examples=10, dim=(1, 10), figsize=(10, 1)):
    noise = np.random.normal(0, 1, (examples, latent_dim))
    fake_images = generator.predict(noise)
    fake_images = 0.5 * fake_images + 0.5  # Escalar de nuevo al rango [0, 1]
    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i + 1)
        plt.imshow(fake_images[i, :, :, 0], cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Entrenar la GAN
d_losses, g_losses = train_gan(gan, generator, discriminator, x_train, latent_dim)

```

```python
# Trama de las curvas de pÃ©rdida del generador y el discriminador
plt.figure(figsize=(12, 6))
plt.plot(d_losses, label="Discriminator Loss")
plt.plot(g_losses, label="Generator Loss")
plt.legend()
plt.xlabel("Ã‰poca")
plt.ylabel("PÃ©rdida")
plt.title("Curvas de PÃ©rdida del Generador y el Discriminador")
plt.show()

# Modificar una dimensiÃ³n especÃ­fica del espacio latente
selected_dimension = 10
noise = np.random.normal(0, 1, (1, latent_dim))
noise[0, selected_dimension] = 21  # Cambiar el valor de la dimensiÃ³n seleccionada
generated_image = generator.predict(noise)


# DespuÃ©s del entrenamiento seleccionar epoca
selected_epoch = 10000
selected_noise = np.random.normal(0, 1, (1, latent_dim))
# Modificar una dimensiÃ³n especÃ­fica del espacio latente
nuevo_valor = 2.0  # Cambiar el valor de la dimensiÃ³n seleccionada
selected_noise_modificado = selected_noise.copy()
selected_noise_modificado[0, selected_dimension] = nuevo_valor  # Modificar la dimensiÃ³n seleccionada

# Generar una imagen con el generador modificado
generated_image_modificado = generator.predict(selected_noise_modificado)

plt.imshow(generated_image[0, :, :, 0], cmap='gray')
plt.axis('off')
plt.title("Imagen generada con dimensiÃ³n modificada")
plt.show()
```
